A jupyter notebook to sample from kernel

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gaussianKernel(x,x_p, t0,t1,t2,t3):
    return t1*np.exp(-0.5*t1 *(x-x_p)**2) +t2 +t3*x*x_p
```

```python
def exponentialKernel(x,xp,t):
    return np.exp(-t*abs(x-xp))
```

```python
x = np.linspace(start=-1, stop=1, num=100)
x_p = 0.5
t0,t1,t2,t3 = 1. , 64., 0,0.
```

# Kernel
## Gaussian Kernel

```python
for x_p in [0.5, -0.5, 0, -0.2]:
    plt.plot(x,gaussianKernel(x,x_p,t0,t1,t2,t3))
```

```python
N = 100
z = np.linspace(-1,1,N)
C = np.zeros([N,N])
for i,xn in enumerate(z):
    for j, xm in enumerate(z):
        C[i, j ] = gaussianKernel(xn,xm,t0,t1,t2, t3)
```

```python
# sample for 4 cases
sample = np.random.multivariate_normal(np.zeros(N), C, 4)
```

```python
plt.plot(z,sample.T)
```

```python
for x_p in [0.5, -0.5, 0, -0.2]:
    plt.plot(x,exponentialKernel(x,x_p,t))
```

```python
N = 100
z = np.linspace(-1,1,N)
C = np.zeros([N,N])
for i,xn in enumerate(z):
    for j, xm in enumerate(z):
        C[i, j ] = exponentialKernel(xn,xm,2)
```

```python
sample = np.random.multivariate_normal(np.zeros(N), C, 4)
```

```python
plt.plot(z,sample.T)
```
